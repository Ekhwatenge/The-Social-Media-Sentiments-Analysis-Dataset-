```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("sentimentdataset.csv")

# Display the first few rows of the dataset
print(df.head())

# Step 1: Map individual sentiments to broader themes
theme_mapping = {
    'Positive': 'Positive',
    'Happiness': 'Positive',
    'Joy': 'Positive',
    'Love': 'Positive',
    'Admiration': 'Positive',
    'Affection': 'Positive',
    'Amusement': 'Positive',
    'Enjoyment': 'Positive',
    'Adoration': 'Positive',
    'Anticipation': 'Mixed',
    'Surprise': 'Mixed',
    'Acceptance': 'Mixed',
    'Calmness': 'Neutral',
    'Neutral': 'Neutral',
    'Confusion': 'Negative',
    'Sadness': 'Negative',
    'Fear': 'Negative',
    'Anger': 'Negative',
    'Disgust': 'Negative',
    'Shame': 'Negative',
    'Bitter': 'Negative',
    'Disappointed': 'Negative'
}

df['Theme'] = df['Sentiment'].map(theme_mapping)

print(df['Theme'].isna().sum())  # Check for NaN values
print(df['Sentiment'].unique())  # Check unique sentiment values

# Step 2: Aggregate data by themes
theme_counts = df['Theme'].value_counts().reset_index()
theme_counts.columns = ['Theme', 'Count']

# Step 3: Visualize the themes (Bar Chart)
plt.figure(figsize=(8, 6))
sns.barplot(data=theme_counts, x='Theme', y='Count', palette='muted')
plt.title('Distribution of Sentiment Themes', fontsize=16)
plt.xlabel('Theme', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.savefig('sentiment_themes_distribution.png')  # Save the plot
plt.show()

```

```python

# Setting a style for plots
sns.set(style="whitegrid", palette="muted")

# Analysis 1: Sentiment Distribution
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Sentiment', palette='deep')
plt.title('Distribution of Sentiments', fontsize=16)
plt.xlabel('Sentiment', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.savefig('sentiment_distribution.png')  # Save the plot
plt.show()

# Pie Chart for Sentiment Distribution
plt.figure(figsize=(8, 6))
sentiment_counts = df['Sentiment'].value_counts()
colors = sns.color_palette('pastel')[0:len(sentiment_counts)]
plt.pie(sentiment_counts, labels=sentiment_counts.index, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart of Sentiment Distribution ', fontsize=16)
plt.savefig('sentiment_distribution_pie.png')  # Save the plot
plt.show()

# Horizontal Bar Chart for Sentiment Distribution
plt.figure(figsize=(10, 6))
sns.barplot(x=sentiment_counts.values, y=sentiment_counts.index, palette='coolwarm')
plt.title('Horizontal Bar Chart of Sentiment Distribution', fontsize=16)
plt.xlabel('Count', fontsize=12)
plt.ylabel('Sentiment', fontsize=12)
plt.savefig('sentiment_distribution_horizontal_bar.png')  # Save the plot
plt.show()

# Percentage-Based Bar Chart for Sentiment Distribution
plt.figure(figsize=(8, 6))
sentiment_percentages = (sentiment_counts / sentiment_counts.sum()) * 100
sns.barplot(x=sentiment_percentages.index, y=sentiment_percentages.values, palette='viridis')
plt.title('Percentage Sentiment Distribution', fontsize=16)
plt.xlabel('Sentiment', fontsize=12)
plt.ylabel('Percentage (%)', fontsize=12)
for index, value in enumerate(sentiment_percentages.values):
    plt.text(index, value + 1, f"{value:.1f}%", ha='center', fontsize=10)
plt.savefig('sentiment_distribution_percentage.png')  # Save the plot
plt.show()


# Analysis 2: Geographical Distribution of Sentiments
plt.figure(figsize=(10, 8))
sns.countplot(data=df, x='Country', hue='Sentiment', palette='dark')
plt.title('Geographical Distribution of Sentiments', fontsize=16)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='Sentiment')
plt.savefig('geographical_sentiment_distribution.png')  # Save the plot
plt.show()

# Analysis 3: Platform Usage by Sentiments
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Platform', hue='Sentiment', palette='pastel')
plt.title('Platform Usage by Sentiments', fontsize=16)
plt.xlabel('Platform', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='Sentiment')
plt.savefig('platform_sentiment_distribution.png')  # Save the plot
plt.show()

# Analysis 4: Engagement Metrics (Likes and Retweets) by Sentiments
engagement_metrics = df.groupby('Sentiment')[['Likes', 'Retweets']].sum().reset_index()
engagement_metrics_melted = engagement_metrics.melt(id_vars='Sentiment', value_vars=['Likes', 'Retweets'], var_name='Metric')

plt.figure(figsize=(10, 6))
sns.barplot(data=engagement_metrics_melted, x='Sentiment', y='value', hue='Metric', palette='coolwarm')
plt.title('Engagement Metrics by Sentiments (Likes and Retweets)', fontsize=16)
plt.xlabel('Sentiment', fontsize=12)
plt.ylabel('Total Engagements', fontsize=12)
plt.legend(title='Metric')
plt.savefig('engagement_metrics_sentiments.png')  # Save the plot
plt.show()

```

```python
# Analysis 5: Temporal Analysis - Sentiments Over Time
df['Date'] = pd.to_datetime(df['Timestamp']).dt.date

sentiments_over_time = df.groupby(['Date', 'Sentiment']).size().reset_index(name='Count')

plt.figure(figsize=(12, 6))
sns.lineplot(data=sentiments_over_time, x='Date', y='Count', hue='Sentiment', palette='viridis')
plt.title('Temporal Analysis of Sentiments Over Time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Count of Posts', fontsize=12)
plt.legend(title='Sentiment')
plt.savefig('temporal_sentiments_analysis.png')  # Save the plot
plt.show()

# Additional Insights: Hashtags Usage Frequency
hashtags = df['Hashtags'].str.split().explode().value_counts().head(10)

plt.figure(figsize=(8, 6))
sns.barplot(x=hashtags.values, y=hashtags.index, palette="mako")
plt.title('Top 10 Most Used Hashtags', fontsize=16)
plt.xlabel('Frequency', fontsize=12)
plt.ylabel('Hashtags', fontsize=12)
plt.savefig('top_hashtags.png')  # Save the plot
plt.show()
```
